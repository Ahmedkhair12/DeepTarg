```python
from sklearn.metrics import roc_curve, precision_recall_curve, confusion_matrix, auc
import matplotlib.pyplot as plt
import numpy as np
```

```python
g_truth_1 = np.load('cross_comparisons/comp_ground_truths_comp1.npy')
pred_1 = np.load('cross_comparisons/predictions_comp1.npy')

g_truth_2 = np.load('cross_comparisons/comp_ground_truths_comp2.npy')
pred_2 = np.load('cross_comparisons/predictions_comp2.npy')

g_truth_3 = np.load('cross_comparisons/comp_ground_truths_comp3.npy')
pred_3 = np.load('cross_comparisons/predictions_comp3.npy')

g_truth_4 = np.load('cross_comparisons/comp_ground_truths_comp4.npy')
pred_4 = np.load('cross_comparisons/predictions_comp4.npy')
```

```python

# Compute ROC curves and AUC scores
fpr1, tpr1, _ = roc_curve(g_truth_1, pred_1)
auc1 = auc(fpr1, tpr1)
fpr2, tpr2, _ = roc_curve(g_truth_2, pred_2)
auc2 = auc(fpr2, tpr2)
fpr3, tpr3, _ = roc_curve(g_truth_3, pred_3)
auc3 = auc(fpr3, tpr3)
fpr4, tpr4, _ = roc_curve(g_truth_4, pred_4)
auc4 = auc(fpr4, tpr4)

# Plotting ROC curves
plt.figure(figsize=(8, 6))

# Plot each ROC curve with a different color
colors = ['blue', 'orange', 'green', 'red']
params = ['64-channels', '128-channels', '256-channels', '512-channels' ]
for i, (fpr, tpr, auc_score, param) in enumerate(zip([fpr1, fpr2, fpr3, fpr4], [tpr1, tpr2, tpr3, tpr4], [auc1, auc2, auc3, auc4], params)):
    plt.plot(fpr, tpr, color=colors[i], lw=2, label=f'{param} (AUC = {auc_score:.3f})')

# Plotting the diagonal line (random classifier)
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')

# Adding labels and title
plt.title('Hyperparameters: (GraphConv, 2-layers, 25-Epochs, LR=0.001)')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
